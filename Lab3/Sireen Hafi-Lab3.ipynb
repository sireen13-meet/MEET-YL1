```python
import csv
f = open("cig.csv", "rb")
r = csv.reader(f)
rownum = 0
weight = []
volume = []

for row in r:
    if rownum != 0:
        a = row[0].split()
        b = map(float, a[1:])
        weight.append(b[1])
        volume.append(b[2])

    rownum += 1

f.close()

from pylab import *

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('The amount of rainfall')
ax.set_xlabel('The year (g)')
ax.set_ylabel('The amount of rainfall(cc)')
scatter(weight, volume)
fig.savefig("rainfall_with_titles.png")
```

```python
import csv
```

```python
f = open("cig.csv", "rb")
r = csv.reader(f)
rownum = 0
weight = []
volume = []

for row in r:
    if rownum != 0:
        a = row[0].split()
        b = map(float, a[1:])
        weight.append(b[1])
        volume.append(b[2])

    rownum +=1
    
print weight
print volume

f.close()
```

```python
from pylab import *
fig = plt.figure()

ax = fig.add_subplot(111)

ax.set_title('sireen title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
```
